```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

from sklearn.datasets import load_iris
```

```python
# iris = load_iris()
X = np.array([[ 5.1,  4.9,  2.4,  2.4],
              [ 4.9,  3 ,  1.9,  3.9]])
# 将目标矩阵中心化
# X = X - X.mean(axis=0)
X = X.T
X = X - X.mean(axis=0)
X
```

进行绘图

```python
plt.figure(figsize=(6, 6))
size = 6
plt.xlim(-size, size)
plt.ylim(-size, size)
plt.plot([-size, size], [0, 0], label = "x_axis", c="blue")
plt.plot([0, 0], [-size, size], label = "y_axis", c="blue")
plt.legend(loc='best')
plt.scatter(X[:, 0], X[:, 1], label='Original data')
plt.show()
```
